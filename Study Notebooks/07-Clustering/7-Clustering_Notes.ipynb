# 1: Overview

Clustering in machine learning is the process of organizing unlabeled data points into suitably similar groups, called clusters. These groups are formed based on similarities between the data points. It's like sorting a pile of socks by color - the algorithm finds socks that are similar (same color) and puts them together.

-   Unsupervised learning: Unlike classification where data points have predefined labels, clustering deals with unlabeled data. The algorithm figures out the similarities on its own.
-   Similarity measure: A distance metric is used to determine how similar data points are. Common examples include Euclidean distance for continuous data and Hamming distance for categorical data.
-   Applications: Clustering has a wide range of applications like customer segmentation in marketing, image compression, and anomaly detection.

Clustering groups similar data points together, enabling  the discovery or identification of hidden patterns or relationships within the data.

Note:

-   Clustering is a form of machine learning where groups, clusters, or classes, are not known ahead of time so groupings are created by looking at similar and shared characteristics among the things being grouped. (homogeneous groups)
-   Given a set of potential class members, the task of clustering is to establish the existence of two or more classes or clusters in the members.
-   These clusters are often determined through trial and error by grouping things together by looking for similar features, analyzing the results to see how good they are, and repeating this learning process until the groups are deemed acceptable.
-   Clustering is an example of unsupervised learning because classes are unknown at the start.

# 3: Application of Clustering

Customer Segmentation
-   Marketing: Businesses use clustering to segment their customers based on purchasing behavior, demographics, or preferences. This helps in targeted marketing and personalized promotions.
-   Retail: Retailers analyze shopping patterns to create segments of customers with similar buying habits, allowing for better inventory management and personalized recommendations.

Document Clustering
-   News Aggregation: Clustering news articles into topics or events helps in organizing and summarizing large volumes of news content.
-   Search Engines: Grouping similar search results together enhances the user's search experience by providing more relevant results.

Image Segmentation
-   Medical Imaging: Clustering is used in segmenting different tissues, organs, or tumors in medical images like MRI or CT scans, aiding in diagnosis and treatment planning.
-   Object Detection: Clustering pixels in images helps in identifying and classifying different objects within the images.

Anomaly Detection
-   Fraud Detection: In finance, clustering is used to identify unusual patterns or transactions that may indicate fraudulent activity.
-   Network Security: Clustering network traffic data can help in detecting anomalies that could signify potential security breaches.

Recommender Systems
-   Entertainment: Streaming services like Netflix or Spotify use clustering to group users with similar tastes and recommend content based on the preferences of these groups.
-   E-commerce: Online stores use clustering to recommend products based on the purchasing behavior of similar customers.

Biological Data Analysis
-   Genomics: Clustering genes or proteins based on their expression levels or structural features helps in understanding biological functions and disease mechanisms.
-   Ecology: Clustering species based on traits or environmental factors aids in studying biodiversity and ecosystem dynamics.

Social Network Analysis
-   Community Detection: Clustering is used to identify communities or groups of users with similar interests or connections in social networks.
-   Influencer Identification: Analyzing clusters of users helps in identifying key influencers or opinion leaders within a network.

Urban Planning
-   Traffic Analysis: Clustering traffic data helps in identifying patterns and hotspots, aiding in traffic management and infrastructure development.
-   Resource Allocation: Clustering residential areas based on various factors such as population density and economic status helps in efficient allocation of resources and services.

Healthcare
-   Patient Segmentation: Clustering patients based on medical history, symptoms, or genetic data helps in personalized treatment and management plans.
-   Disease Outbreak Detection: Clustering cases based on geographic and temporal data can help in early detection and control of disease outbreaks.

Financial Market Analysis
-   Stock Market Segmentation: Clustering stocks based on their price movements and other financial metrics helps in identifying groups of stocks with similar behavior, aiding in portfolio management and investment strategies.
-   Risk Management: Clustering financial instruments based on risk profiles aids in better risk assessment and management.

# 4: Classification vs Clustering

Classification and clustering are techniques used in machine learning to organize data into groups, but they differ in how they achieve this organization:

-   Classification: This is a supervised learning technique. It involves using labeled data, where each data point has a predefined class label. The machine learning algorithm learns from this labeled data to classify new, unlabeled data points. For instance, an algorithm can be trained on emails labeled as "spam" or "not spam" to classify new incoming emails.

-   Clustering: This is an unsupervised learning technique. Unlike classification, clustering deals with unlabeled data. The algorithm identifies similarities between data points and groups them accordingly. These groups, called clusters, are formed based on the inherent patterns or characteristics within the data itself. For example, a clustering algorithm can group customers based on their purchase history, revealing shopping habits or demographics.

# 5: KMeans Clustering

## Overview

k-means clustering is a popular unsupervised machine learning algorithm used for partitioning data into predefined groups (clusters).

The goal of k-means clustering is to group similar data points together, minimizing the within-cluster variance (how spread out the data points are within a cluster).

## The basic process of how k-means clustering works is as follows:
-   Define the number of clusters (k): This is a crucial parameter, and choosing the right k can significantly impact the results. There are techniques to help you decide on the optimal k, but it often involves some experimentation.
-   Initialize centroids: These are the initial representatives for each cluster, typically chosen randomly from the data points. You can think of them as the "centers of gravity" for each cluster.
-   Assign data points to clusters: Each data point is assigned to the closest centroid based on a distance metric (often Euclidean distance).
-   Recalculate centroids: Once all points are assigned, the centroid for each cluster is recalculated as the average of all the points belonging to that cluster.
-   Repeat: Steps 3 and 4 are repeated iteratively. In each iteration, data points might switch clusters based on the updated centroids. This continues until the centroids no longer change significantly, indicating convergence (finding the optimal cluster configuration).

Note:
-   K-means is a centroid-based algorithm, relying on the concept of centroids to represent each cluster.
-   It minimizes the sum of squared distances between data points and their assigned cluster centroid.
-   The initial placement of centroids can influence the final clustering results.
-   k-means works best with numerical data and assumes spherical clusters (data points tend to be evenly distributed around the centroid).

## 5.1: KMeans Process

### Euclidean Distance

In 2-dimensions, how do we measure the "distance" between 2 points? Euclidean distance refers to the straight-line distance between two points in space. It's the shortest distance imaginable, like stretching a tight string between them. (See *Measuring Distances for details.)

### Centroid

The centroid of a collection of points refers to the average location of all those points. It essentially acts like the center of gravity for the group.

Imagine balancing a flat cutout with all the points scattered across it. The centroid would be the exact point where you could place a pin to perfectly balance the cutout – it represents the point where all the "weight" of the individual points is evenly distributed.
-   In 1 dimension (a line): The centroid would be the average value of all the points on the line.
-   In 2 dimensions (a flat plane): The centroid would be the point with the average x-coordinate and average y-coordinate of all the points in the plane.
-   In 3 dimensions (like a cube): The centroid would be the point with the average x, y, and z coordinates of all the points.

In k-means clustering, centroids represent the center of each cluster, and data points are assigned to the cluster with the closest centroid. This helps group similar data points together based on their "average" location.

## 5.2: Determining k in kmeans clustering

An important step for any unsupervised clustering algorithm is to determine the optimal number of clusters into which the data may be clustered. We need to have some methods that can help us decide the best number of clusters.

The k-means clustering algorithm requires us to initialize the  k
clusters and we can iteratively adjust these k clusters until till the k-centroids reach an equilibrium state. 

### Scatter Plots with Contours

For simple datasets with a few number of variables, one can often determine the number of clusters through simple visualization of the data. For example, from the following scatter plot we can clearly distinguish the 4 clusters.

### Elbow Curve Method

The elbow curve method is a popular approach to determining the optimal number of clusters.  In this method, we iterate from k=1
to some upper limit for the number of possible clusters, k=N. For each value of k, we calculate the within-cluster sum of squares (WCSS) value. WCSS is defined as the sum of the square distance between points in a cluster and the cluster centroid. This is also known as the distortion score. As k increases, the WCSS score decreases. Initially the score drops off dramatically, but at some point, the score will start to drop off more slowly with k.

<div>
    <img src="../src/images/kmeanselbow.png" width="600">
</div>

The k-value point where the rate of change of WCSS changes is referred to as an elbow. The value of k at the elbow gives an indication of the suitable choice of k.

Note: The elbow method gives a good initial guess for k. But, it may not be the best choice.

### Silhouette Analysis

Silhouette analysis can be used to study the separation distance between the resulting clusters. The silhouette plot displays a measure of how close each point in one cluster is to points in the neighboring clusters and thus provides a way to assess parameters like number of clusters visually. This measure has a range of [-1, 1].

How to analyze the Silhouette plot to select an optimal n_cluster.
-   A sub-optimal n_cluster will show the presence of clusters below the average silhouette score. For this given data, none of the n_clusters shows value below the average score. (The average score is indicated with the red-dotted line).
-   A sub-optimal n_cluster will show wide fluctuation in the size of the silhouette plot. For this given data, a wide fluctuation is seen in almost all the n_clusters , although less in the last where the number of values of n_cluster is five.

Thus, our preferred n_cluster will be five, as it passes the two criteria.
