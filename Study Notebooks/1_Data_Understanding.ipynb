# 1) Data Coming In

lets use Mall_Customers.csv

```python
# IMPORT ALL PACKAGES
import pandas as pd
import numpy as np
import os
import sys
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
df = pd.read_csv('src/data/Mall_Customers.csv')

# Preview the dataset
df.head()
```

## 2) Data Understanding

```python
# Get basic information about the dataset
print ("Dataset Information:")
df.info()

# Check for missing values with column names and counts
print("\nMissing Values:")
df.isnull().sum()
```

```python
df.describe()
```

## 3) Distributions

```python
# Gender distribution
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
# Age distribution
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Scatter plot of Spending Score vs Annual Income, colored by Gender
sns.scatterplot(data=df, x='Annual_Income', y='Spending_Score', hue='Gender')
plt.title("Spending Score vs Annual Income")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```
